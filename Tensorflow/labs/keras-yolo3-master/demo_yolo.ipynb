# YOLO模块单元测试历程

## letterbox_image
    不改变图像长宽比例，取新图像的长与旧图像的长，新图像的宽与旧图像的宽比之最小值为比例重设图像大小，居中，
    空白处以(128,128,128)填充.

```python
%matplotlib inline
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from yolo3.utils import letterbox_image

img_file='%s/work/data/yolo/1.jpg'%os.getenv('HOME')
target_size=(400,200) #(width,height)
#读取图像
img_src=Image.open(img_file)
#图像处理
img_dst=letterbox_image(img_src,target_size)

#图像显示
print('img_src.size:',img_src.size)
print('img_dst.size:',img_dst.size)
plt.subplot(1,2,1)
plt.imshow(img_src)
plt.title('src')
plt.subplot(1,2,2)
plt.imshow(img_dst)
plt.title('dst')
```
