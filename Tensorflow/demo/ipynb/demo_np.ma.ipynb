## np.ma.compress_rowcols
 取消二维数组中包含掩码值的行和/或列

```python
x = np.ma.array(np.arange(9).reshape(3, 3), mask=[[1, 0, 0],
                                                   [1, 0, 0],
                                                   [0, 0, 0]])
print('x:')
print(x)
print('\nx.mask:')
print(x.mask)
print('\nnp.ma.compress_rowcols(x):')
print(np.ma.compress_rowcols(x))
print('\nnp.ma.compress_rowcols(x, 0):')
print(np.ma.compress_rowcols(x, 0))
print('\nnp.ma.compress_rowcols(x, 1):')
print(np.ma.compress_rowcols(x, 1))
```

## np.ma.masked_invalid
屏蔽nan数据

```python
x = np.ma.array(
    [[1,2,3],
     [np.nan,5,6],
     [7,8,9]
    ]
    , mask=[[1, 0, 0],
            [0, 0, 0],
            [0, 0, 0]])

print('x:')
print(x)
print('\n np.ma.masked_invalid(x):')
print(np.ma.masked_invalid(x))
print('np.ma.compress_rows(np.ma.masked_invalid(x)):')
print(np.ma.compress_rows(np.ma.masked_invalid(x)))
```
