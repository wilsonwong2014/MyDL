## 轮廓查找
    
* 阈值处理

    Python: cv2.threshold( src, thresh, maxval, type [ , dst ] ) → retval, dst
    
    参数:
        src ------ 阈值处理的灰度图
        thresh --- 阈值
        maxval --- 阈值设值，参考手册
        type ----- 阈值处理方法：
            THRESH_BINARY
            THRESH_BINARY_INV
            THRESH_TRUNC
            THRESH_TOZERO
            THRESH_TOZERO_INV

    返回值:
        retval --- thresh
        dst ------ 阈值处理后的结果
        
* 轮廓查找

    Python: cv2.findContours( image, mode, method [ , contours [ , hierarchy [ , offset ]]] )
        → contours, hierarchy
        
    参数:
        image -- 寻找轮廓的图像；
        mode  -- 表示轮廓的检索模式，有四种（本文介绍的都是新的cv2接口）：
            cv2.RETR_EXTERNAL表示只检测外轮廓
            cv2.RETR_LIST检测的轮廓不建立等级关系
            cv2.RETR_CCOMP建立两个等级的轮廓，上面的一层为外边界，里面的一层为内孔的边界信息。如果内孔内还有一个连通物体，这个物体的边界也在顶层。
            cv2.RETR_TREE建立一个等级树结构的轮廓。
        method--轮廓的近似办法
            cv2.CHAIN_APPROX_NONE存储所有的轮廓点，相邻的两个点的像素位置差不超过1，即max（abs（x1-x2），abs（y2-y1））==1
            cv2.CHAIN_APPROX_SIMPLE压缩水平方向，垂直方向，对角线方向的元素，只保留该方向的终点坐标，例如一个矩形轮廓只需4个点来保存轮廓信息
            cv2.CHAIN_APPROX_TC89_L1，CV_CHAIN_APPROX_TC89_KCOS使用teh-Chinl chain 近似算法        
            
    返回值:
        contours -- list,每个元素表示一个轮廓，用np.array表示
        hierarchy-- 返回一个可选的hiararchy结果，这是一个ndarray，其中的元素个数和轮廓个数相同，每个轮廓contours[i]对应4个hierarchy元素hierarchy[i][0] ~hierarchy[i][3]，分别表示后一个轮廓、前一个轮廓、父轮廓、内嵌轮廓的索引编号，如果没有对应项，则该值为负数。
        
* 轮廓绘制

    Python: cv2.drawContours( image, contours, contourIdx, color [ , thickness [ , lineType [ , hierarchy [ ,
maxLevel [ , offset ]]]]] ) → None



```python
import os
import cv2  
import matplotlib.pyplot as plt

img_file='%s/work/data/gtest/1.jpg'%os.getenv('HOME')                                #图像文件
img = cv2.imread(img_file)                                                           #读取图像
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)                                          #转灰度图
ret, binary = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)                          #阈值处理
contours, hierarchy = cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE) #轮廓查找  
cv2.drawContours(img,contours,-1,(0,0,255),2)                                        #轮廓绘制
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))                                      #显示结果
```

## 轮廓查找,多边形拟合,凸包计算
* 多边形拟合

    Python: cv2.approxPolyDP( curve, epsilon, closed [ , approxCurve ] ) → approxCurve
    
    参数
        curve ----轮廓
        epsilon --输出精度，最大边长距离
        closed ---是否闭合
    返回值
        approxCurve -- 拟合多边形轮廓
         

```python
'''
轮廓查找,多边形拟合,凸包计算
'''

import os;
import sys;


import cv2
import numpy as np

img_file='%s/work/data/gtest/hammer.jpg'%os.getenv('HOME')

img = cv2.pyrDown(cv2.imread(img_file, cv2.IMREAD_UNCHANGED)); #金字塔图像缩小

ret, thresh = cv2.threshold(cv2.cvtColor(img.copy(), cv2.COLOR_BGR2GRAY) , 127, 255, cv2.THRESH_BINARY); #阈值处理
black = cv2.cvtColor(np.zeros((img.shape[1], img.shape[0]), dtype=np.uint8), cv2.COLOR_GRAY2BGR); #灰度图转BGR

contours, hier = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE);     #轮廓查找

for cnt in contours:
  epsilon = 0.01 * cv2.arcLength(cnt,True);     #cv2.argLength-计算弧长
  approx = cv2.approxPolyDP(cnt,epsilon,True);  #多边形拟合
  hull = cv2.convexHull(cnt);                   #计算凸包

  cv2.drawContours(black, [cnt], -1, (255, 0, 0), 2)     #蓝色-原始轮廓
  cv2.drawContours(black, [approx], -1, (0, 255, 0), 2)  #绿色-多边形拟合
  cv2.drawContours(black, [hull], -1, (0, 0, 255), 2)    #红色-凸包

#结果显示    
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(black,cv2.COLOR_BGR2RGB))
```

## 傅里叶变换

```python
'''
傅里叶变换
'''

import os;
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_file='%s/work/data/gtest/1.jpg'%os.getenv('HOME') #图像路径

img = cv2.imread(img_file, 0);                        #读取图像
f = np.fft.fft2(img);                                 #傅里叶变换 
fshift = np.fft.fftshift(f);                          #频谱居中
magnitude_spectrum = 20 * np.log(np.abs(fshift));     #频谱对数处理

#高通
row, cols = img.shape
crow, ccol = int(row / 2), int(cols / 2)
fshift[crow - 30: crow+30, ccol - 30: ccol + 30] = 0

#频谱居中复原
f_ishift = np.fft.ifftshift(fshift)
#傅里叶逆变换
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)

#原始图
plt.subplot(131), plt.imshow(img, cmap = "gray")
plt.title("Input"), plt.xticks([]), plt.yticks([])

#图像频谱
plt.subplot(132), plt.imshow(magnitude_spectrum, cmap = "gray")
plt.title('magnitude_spectrum'), plt.xticks([]), plt.yticks([])

#高通
plt.subplot(133), plt.imshow(img_back, cmap = "gray")
plt.title("Input in JET"), plt.xticks([]), plt.yticks([])
plt.show()
```

## 霍夫线变换

    cv2.HoughLinesP()函数原型：
    HoughLinesP(image, rho, theta, threshold, lines=None, minLineLength=None, maxLineGap=None) 
      image：        必须是二值图像，推荐使用canny边缘检测的结果图像； 
      rho:           线段以像素为单位的距离精度，double类型的，推荐用1.0 
      theta：        线段以弧度为单位的角度精度，推荐用numpy.pi/180 
      threshod:      累加平面的阈值参数，int类型，超过设定阈值才被检测出线段，值越大，基本上意味着检出的线段越长，
                     检出的线段个数越少。根据情况推荐先用100试试
      lines：        这个参数的意义未知，发现不同的lines对结果没影响，但是不要忽略了它的存在 
      minLineLength：线段以像素为单位的最小长度，根据应用场景设置 
      maxLineGap：   同一方向上两条线段判定为一条线段的最大允许间隔（断裂），超过了设定值，则把两条线段当成一条线段，
                     值越大，允许线段上的断裂越大，越有可能检出潜在的直线段


```python
'''
霍夫线变换

cv2.HoughLinesP()函数原型：
HoughLinesP(image, rho, theta, threshold, lines=None, minLineLength=None, maxLineGap=None) 
  image：        必须是二值图像，推荐使用canny边缘检测的结果图像； 
  rho:           线段以像素为单位的距离精度，double类型的，推荐用1.0 
  theta：        线段以弧度为单位的角度精度，推荐用numpy.pi/180 
  threshod:      累加平面的阈值参数，int类型，超过设定阈值才被检测出线段，值越大，基本上意味着检出的线段越长，
                 检出的线段个数越少。根据情况推荐先用100试试
  lines：        这个参数的意义未知，发现不同的lines对结果没影响，但是不要忽略了它的存在 
  minLineLength：线段以像素为单位的最小长度，根据应用场景设置 
  maxLineGap：   同一方向上两条线段判定为一条线段的最大允许间隔（断裂），超过了设定值，则把两条线段当成一条线段，
                 值越大，允许线段上的断裂越大，越有可能检出潜在的直线段
'''

import os;
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_file='%s/work/data/gtest/lines.jpg'%os.getenv('HOME') #图像路径

img = cv2.imread(img_file);                 #读取原始图像
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY);#灰度图转换
edges = cv2.Canny(gray,50,120);             #canny边缘检测
minLineLength = 20;                         #线段最小长度
maxLineGap = 5;                             #空隙阈值
lines = cv2.HoughLinesP(edges,1,np.pi/180,50,minLineLength,maxLineGap);#霍夫线检测,返回:nx1x4
for x in lines:
  x1,y1,x2,y2=x[0,0],x[0,1],x[0,2],x[0,3];
  cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2);

plt.subplot(1,2,1)
plt.imshow(edges)
plt.subplot(1,2,2)
plt.imshow(img)
```

## 霍夫圆变换

```python
'''
霍夫圆变换
'''

import os;
import sys;
import matplotlib.pyplot as plt

import cv2
import numpy as np

img_file='%s/work/data/gtest/planet_glow.jpg'%os.getenv('HOME') #图像路径

planets = cv2.imread(img_file);                      #读取原始图像
gray_img = cv2.cvtColor(planets, cv2.COLOR_BGR2GRAY);#灰度图转换
img = cv2.medianBlur(gray_img, 5);                   #中值滤波
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR);         #灰度图转BGR

#霍夫圆变换:返回3xn数组,第一行:x,第二行:y,第三行:r
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,120,
                            param1=100,param2=30,minRadius=0,maxRadius=0)

circles = np.uint16(np.around(circles))

for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(planets,(i[0],i[1]),i[2],(0,255,0),2); #霍夫圆
    # draw the center of the circle
    cv2.circle(planets,(i[0],i[1]),2,(0,0,255),3);    #圆心

plt.imshow(cv2.cvtColor(planets,cv2.COLOR_BGR2RGB))
```

## 高通滤波/高斯滤波

```python
'''
高通滤波
'''

import os;

import cv2
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt

kernel_3x3 = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

kernel_5x5 = np.array([[-1, -1, -1, -1, -1],
                       [-1,  1,  2,  1, -1],
                       [-1,  2,  4,  2, -1],
                       [-1,  1,  2,  1, -1],
                       [-1, -1, -1, -1, -1]])

img_file='%s/work/data/gtest/pycv_images/statue_small.jpg'%os.getenv('HOME') #图像路径

img = cv2.imread(img_file, 0);              #读取原始图像
k3 = ndimage.convolve(img, kernel_3x3);     #卷积
k5 = ndimage.convolve(img, kernel_5x5);     #卷积

blurred = cv2.GaussianBlur(img, (17,17), 0);#高斯滤波
g_hpf = img - blurred

dpi=72. #fig.dpi => 72
fig=plt.figure(figsize=(1024/dpi,1024/dpi))
plt.subplot(1,5,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))    ,plt.title('img-src')
plt.subplot(1,5,2)
plt.imshow(cv2.cvtColor(k3,cv2.COLOR_BGR2RGB))     ,plt.title('conv-3x3')
plt.subplot(1,5,3)
plt.imshow(cv2.cvtColor(k5,cv2.COLOR_BGR2RGB))     ,plt.title('conv-5x5')
plt.subplot(1,5,4)
plt.imshow(cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB)),plt.title('gauss')
plt.subplot(1,5,5)
plt.imshow(cv2.cvtColor(g_hpf,cv2.COLOR_BGR2RGB))  ,plt.title('img-gauss')
```

## 基本图形绘制
    绘制直线
    Python: cv2.line( img, pt1, pt2, color [ , thickness [ , lineType [ , shift ]]] ) → None
    绘制矩形
    Python: cv2.rectangle( img, pt1, pt2, color [ , thickness [ , lineType [ , shift ]]] ) → None
    绘制椭圆
    Python: cv2.ellipse( img, center, axes, angle, startAngle, endAngle, color [ , thickness [ , lineType [ , shift ]]] ) → None
    Python: cv2.ellipse( img, box, color [ , thickness [ , lineType ]] ) → None    
    绘制圆形
    Python: cv2.circle( img, center, radius, color [ , thickness [ , lineType [ , shift ]]] ) → None
    绘制多边形
    Python: cv2.polylines( img, pts, isClosed, color [ , thickness [ , lineType [ , shift ]]] ) → None
    填充多边形
    Python: cv2.fillPoly( img, pts, color [ , lineType [ , shift [ , offset ]]] ) → None    
    添加
    Python: cv2.putText( img, text, org, fontFace, fontScale, color [ , thickness [ , lineType [ , bottomLeftOrigin ]]] ) → None
    

```python
#基本图形绘制
#===============
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

w=800 #宽度
h=600 #高度

#创建空白图像
img=np.zeros((h,w,3),dtype=np.uint8)
#绘制直线的line函数
cv2.line(img=img,pt1=(10,10),pt2=(200,300),color=(255,0,0))
#绘制矩形的rectangle函数
cv2.rectangle(img=img,pt1=(50,100),pt2=(200,300),color=(0,255,0))
#绘制椭圆的ellipse函数
cv2.ellipse(img=img,center=(256,256),axes=(100,50),angle=45,startAngle=0,endAngle=180,color=(20,213,79),thickness=-1)
#绘制圆的circle函数
cv2.circle(img=img,center=(100,100),radius=50,color=(0,0,255))
#绘制填充多边形的fillPoly函数
pts=np.array([[400,50],[600,50],[600,300],[400,300]],np.int32) #数据类型必须是int32 
pts=pts.reshape((-1,1,2))
'''这里 reshape 的第一个参数为-1, 表明这一维的长度是根据后面的维度的计算出来的。
   如果第三个参数是 False，我们得到的多边形是不闭合的（首尾不相连）。
'''
cv2.fillPoly(img,[pts],color=(0,255,255))
cv2.polylines(img,[pts],True,(0,0,255),1) # 图像，点集，是否闭合，颜色，线条粗细
#添加文本
cv2.putText(img,text="OpenCV",org=(80,90),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=2,color=(255,255,255),thickness=3)

dpi=72
fig=plt.figure(figsize=(w/dpi,h/dpi))
plt.imshow(img)
```

## 在图像上绘制矩形框序列

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#在图像上绘制图形
def draw_boxs(img,boxs,thickness,texts=None,colors=None):
    '''
    img.shape => (height,width,channels)
    boxs      =>  [[x1,y1,x2,y2],......]
    thickness =>  int
    texts     =>  ['text1','text2',....]
    colors    =>  [(255,0,0),..........]
    '''
    for i,box in enumerate(boxs):
        pt1=(box[0],box[1])
        pt2=(box[2],box[3])
        text='' if texts is None else texts[i]
        color=(255,0,0) if colors is None else colors[i]
        cv2.rectangle(img=img,pt1=pt1,pt2=pt2,color=color,thickness=thickness)
        cv2.putText(img,text=text,org=pt1,fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=1,color=color,thickness=thickness)
    return img


#---test----
img=np.zeros((600,800,3),dtype=np.uint8)
boxs=[[50,100,300,200],[100,50,200,300]] #(x1,y1,x2,y2)
img=draw_boxs(img,boxs,thickness=2,texts=['a','b'],colors=[(255,0,0),(0,255,0)])
boxs=[[60,130,330,230],[120,60,280,390]] #(x1,y1,x2,y2)
img=draw_boxs(img,boxs,thickness=2,texts=['a','b'],colors=[(255,0,0),(0,255,0)])
#plt.figure(figsize=(800/72.,600/72.))
plt.imshow(img)
```

## 提取感兴趣区域ROI
    1- 建立ROI多边形np.array([],np.int32),并使用该ROI,通过cv2.fillPoly()或cv2.fillConvexPoly()建立蒙版mask
    2- 使用cv2.bitwise_and()等建立mask
    
    Python: cv2.bitwise_and( src1, src2 [ , dst [ , mask ]] ) → dst

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

img_file='%s/work/data/gtest/1.jpg'%os.getenv('HOME')                      #图像路径
img=cv2.imread(img_file)                                                   #读取图像
ROI=np.zeros(img.shape,dtype=np.uint8)                                     #创建掩码模板
pts=np.array([[10,10],[100,10],[100,100],[50,150],[10,100]],dtype=np.int32)#多边形区域ROI
pts=pts.reshape([-1,1,2])                                                  #reshape
cv2.fillPoly(ROI,[pts],color=(255,255,255))                                #建立掩膜
img_roi=cv2.bitwise_and(ROI,img)                                           #提取ROI区域

#显示图像
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(ROI,cv2.COLOR_BGR2RGB))
plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(img_roi,cv2.COLOR_BGR2RGB))                        
```

## 图像融合
    Python: cv2.addWeighted( src1, alpha, src2, beta, gamma [ , dst [ , dtype ]] ) → dst

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

img1_path='%s/work/data/gtest/1.jpg'%os.getenv('HOME')
img2_path='%s/work/data/gtest/2.jpg'%os.getenv('HOME')
img1 = cv2.imread(img1_path) 
img2 = cv2.imread(img2_path) 
h, w, _ = img1.shape 

img2 = cv2.resize(img2, (w,h), interpolation=cv2.INTER_AREA)  # 函数要求两张图必须是同一个size 
img_add = cv2.addWeighted(src1=img1, alpha=0.7, src2=img2, beta=0.5, gamma=0) #图像融合

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB))
plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(img_add,cv2.COLOR_BGR2RGB))
```
